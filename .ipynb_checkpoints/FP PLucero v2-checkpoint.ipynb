### Se consideran dos maneras de resolver el método de Newton:
### Vía sistemas lineales: $J(x_i)\Delta x = -F(x_i)$
###  Vía matriz inversa: $\Delta x=-J(x_i)^{-1}F(x_i)$
#### Definimos doolittle para matrices 2x2 con fin de probar ambos planteamientos de la ec. del método de Newton 

```python
def doolittle2x2(A, b):
    u11, u12 = A[0]
    l21 = A[1][0] / u11
    u22 = A[1][1] - u12*l21
    
    y1 = b[0]
    y2 = b[1] - l21 * y1
    
    x2 = y2/u22
    x1 = (y1 - u12*x2)/u11
    
    return x1, x2
```

# Problema 1
## Ecuaciones de Control de Robots
## $\begin{cases}
x_1^{2} + cos(y) - 2 = 0 \\
e^{x_1} + y - 1 = 0 \\
\end{cases}$

### Por simplicidad, se considera $y$ como $x_2$

# Definimos nuestras F(x) para crear el vector
## $f_1 = x_1^{2} + cos(x_2) - 2$
## $f_2 = e^{x_1} + x_2 - 1$
## $F(X) = 
\left(\begin{array}{cc} 
x_1^{2} + cos(x_2) - 2\\
e^{x_1} + x_2 - 1
\end{array}\right)
$

# Definimos las derivadas parciales para J(x)
## $\dfrac{∂f_1}{∂x_1} = 2x_1 \space\space \dfrac{∂f_1}{∂x_2} = -sin(x_2)$

## $\dfrac{∂f_2}{∂x_1} = e^{x_1} \space\space \dfrac{∂f_2}{∂x_2} = 1$    
        
## $
J(x_i) = 
\left(\begin{array}{cc} 
2x_1 & -sin(x_2)\\
e^{x_1} & 1
\end{array}\right)
$

# La ecuación general para $Δx_i$ es:
## $
\left(\begin{array}{cc} 
2x_1 & -sin(x_2)\\
e^{x_1} & 1
\end{array}\right)
\left(\begin{array}{cc} 
Δx_1 \\
Δx_2
\end{array}\right)
=
\left(\begin{array}{cc} 
x_1^{2} + cos(x_2) - 2\\
e^{x_1} + x_2 - 1
\end{array}\right)
$

### Definimos cada $x_i$ y el vector $X_0$
## $x_1 = 1; x_2 = 1$
## $X_0 = (1, 1)$

# La ec. vectorial planteada para $X_0$ es:
## $
\left(\begin{array}{cc} 
2 & -sin(1)\\
e & 1
\end{array}\right)
\left(\begin{array}{cc} 
Δx_1 \\
Δx_2
\end{array}\right)
=
\left(\begin{array}{cc} 
cos(1) - 1\\
e
\end{array}\right)
$
# Pasando a los sistemas lineales tenemos:
# $\begin{cases}
2Δx_1 - Δx_2sin(1) = cos(1) - 1 \\
eΔx_1 + Δx_2 = e \\
\end{cases}$

## Se emplea un vector $ε = (ε_1, ε_2)$ para almacenar errores relativos, con cada elemento definido como: $ε_i = \biggl|\dfrac {Δx_i}{x_i}\biggr|$

```python
def problema_1(X, imprimir): # Parámetros → Condiciones iniciales
    x1, x2 = X
    E = [100, 100]
    E_tracking = []
    print("")
    print("Condiciones Iniciales: ")
    print(f"x1: {round(x1, 10)}, x2: {round(x2, 10)}")
    print("")

    for a in range(100):

        if all(e < 0.001 for e in E):
            print(a+1, "iteraciones para todo error RP < 0.001%")
            break
        if imprimir:
            print("Iteración:", a)

        # Buscamos obtener: J(x_i)*delta_x=-F(x_i)
        # Conseguimos -F(x_i)
        F1 = x1**2 + cos(x2) - 2
        F2 = e**x1 + x2 - 1

        # Detalle: aquí cada una entra negativa
        # porque el método iguala a -F(x_i)
        F = [-F1, -F2] 

        # Obtenemos J(x_i)
        dF1dx1 = 2*x1
        dF1dx2 = -sin(x2)

        dF2dx1 = e**x1
        dF2dx2 = 1

        J = [ [dF1dx1, dF1dx2],
              [dF2dx1, dF2dx2] ]

        # Resolvemos sistema lineal vía Doolittle para
        # obtener delta_x
        d1, d2 = doolittle2x2(J, F)
        D = [d1, d2]


        # Cálculo de Errores
        for i in range(len(D)):
            E[i] = abs(D[i]/X[i]) * 100

        # Ajustamos este cambio en las x ya que calculamos una diferencia
        x1 += d1
        x2 += d2
        X = [x1, x2]
        
        E_tracking.append(round(sum(np.array(E)), 10))
        if imprimir:
            print(f"x1: {round(x1, 10)}, x2: {round(x2, 10)}")
            print("Σε abs.:", E_tracking[-1])
            print("")

        
    print("")
    print("Evaluación de resultados encontrados:")
    F1 = x1**2 + cos(x2) - 2
    F2 = e**x1 + x2 - 1
    print(f"f1: {round(F1, 10)}, f2: {round(F2, 10)}")
    
    return E_tracking
```

```python
bi_E = problema_1((1,1), True)
```

## ¿Qué ocurre? 
### Las varibles aparentan comportarse de manera semi-errática hasta la iteración 10. Esto puede ser resultado de un atributo del método de Newton: este es extremadamente susceptible a condiciones iniciales (Lange, 2013).
### Tomaremos una condición inicial más adecuada y comparaemos el comportamiento de Σε por iteración

```python
gi_E = problema_1((-5,1), True)
```

```python
# rellenamos con 0's para que gi_E sea del mismo tamaño que bi_E
gi_E.extend([0]*(len(bi_E)-len(gi_E)))

iteracion = np.arange(len(bi_E))
fig, axes = plt.subplots(1, 2, figsize=(20, 9))
fig.suptitle('Resultados de Σε para cada ejecución', fontsize=20)

axes[0].plot(iteracion, bi_E, marker='o', linestyle='--', c="green")
axes[0].set_title("Cambio en Σε por iteración", fontsize=20)
axes[0].set_xlabel("Iteración", fontsize=20)
axes[0].set_ylabel("Σε", fontsize=20)
axes[0].legend(["x1 = 1; x2 = 1"])
# axes[0].set_ylim(0, max_y_value)

axes[1].plot(iteracion, bi_E, marker='o', linestyle='--', c="green")
axes[1].plot(iteracion, gi_E, marker='o', linestyle='--', c="blue")
axes[1].set_title("Comparación en cambio de Σε por iteración", fontsize=20)
axes[1].set_xlabel("Iteración", fontsize=20)
axes[1].set_ylabel("Σε", fontsize=20)
axes[1].legend(["x1 = 1; x2 = 1", "x1 = -5; x2 = 1"])
# axes[0].set_ylim(0, max_y_value)

plt.show()
```

## Para dar explicación a los picos que vemos con las condición X0=(1, 1), debemos primero entender que Σε se refiere a qué tanto se están cambiando las variables en una determinada iteración. 
## Lógicamente, al llegar a una solución del sistema, el "error" se mantendrá en 0, porque las variables no seguirán cambiando.
## Un pico en la gráfica implica un cambio brusco en las variables con respecto a el valor anterior de las mismas, recordar: $ε_i = \biggl|\dfrac {Δx_i}{x_i}\biggr|$. 
## Por tanto, se puede explicar este comportamiento como el método llegando a un vector $X_i$ con valores que "pivotan" la función para que tienda a 0 (un punto de pivote).
## Estos puntos de pivote tienen mucha similitud con los puntos de pivote encuentrados en las superficies de solución de las funciones (Malihoutsaki, et al., 2009).

## Al graficar ambas curvas es fácil apreciar cuales son sus intersecciones. Al plantear una recta entre ambas, encontramos un equivalente al plano planteado por Malihoutsaki & Grapsa.
## Compararemos puntos en esta recta con la hipótesis de encontrar ε con tendencias estrictas hacia 0
## Tomamos un intervalo en $x_1$ arbitrario relativamente cercano a los puntos de intersección ya conocidos: [-3.5, 2.5] y evaluamos la recta entre las intersecciones con estos valores

```python
recta_pivotes = lambda x: -1.7322*x - 1.241
pruebas_x1 = np.arange(-3.5, 3.5, step=1)
pruebas_x2 = recta_pivotes(pruebas_x1)
pruebas_pivotes = [[x1, x2] for x1, x2 in zip(pruebas_x1, pruebas_x2)]
```

```python
print("x1:", pruebas_x1)
print("x2:", pruebas_x2)
```

### Creamos un arreglo con Σε por iteración de cada punto $P=(x_1, x_2)$ evaluado

```python
pivotes_E = []
for prueba in pruebas_pivotes:
    pivotes_E.append(problema_1(prueba, False))
```

```python
iteraciones = [np.arange(len(prueba)) for prueba in pivotes_E]

fig = plt.figure(figsize=(16, 8))

for i in range(len(pivotes_E)):
    plt.plot(iteraciones[i], pivotes_E[i], marker='o', linestyle='--')
plt.title("Cambio en Σε por iteración de de cada ejecución sobre x1 iniciales en [-3.5, 2.5]", fontsize=20)
plt.xlabel("Iteración", fontsize=20)
plt.ylabel("Σε", fontsize=20)
# PONER LABEL CON CADA VALOR INICIAL
# plt.legend(["x1 = 1; x2 = 1"])
plt.legend([f"x1 = {punto[0]}; x2 = {round(punto[1], 3)}" for punto in pruebas_pivotes], fontsize=16)

plt.show()
```

### Existe mucho para ahondar dentro de este problema, si el intérvalo de x1 se expande, la hipótesis propuesta es nula, mostrando que la manera de considerar los puntos de pivote es otra. Sin mencionar que no es factible intentar iniciar usando una recta que se arma con puntos obtenidos como parte de la ejecución

# Problema 2
## Dinámica de Ecosistemas
## $\begin{cases}
x_1log(x_2) + x_2x_3 = 3 \\
x_1^2 + sen(x_2) - x_3 = 1 \\
e^{x_1x_2} + x_3^2 = 4
\end{cases}$

# Definimos nuestras F(x) para crear el vector
## $f_1 = x_1log(x_2) + x_2x_3 - 3$
## $f_2 = x_1^2 + sen(x_2) - x_3 - 1$
## $f_3 = e^{x_1x_2} + x_3^2 - 4$

## $F(X) = 
\left(\begin{array}{cc} 
x_1log(x_2) + x_2x_3 - 3 \\
x_1^2 + sen(x_2) - x_3 - 1 \\
e^{x_1x_2} + x_3^2 - 4 
\end{array}\right)
$

# Definimos las derivadas parciales para J(x)
## $\dfrac{∂f_1}{∂x_1} = log(x_2) \space\space \dfrac{∂f_1}{∂x_2} = \dfrac {x_1}{x_2} + x_3 \space\space \dfrac{∂f_1}{∂x_3} = x_2$ 
## $\dfrac{∂f_2}{∂x_1} = 2x_1 \space\space \dfrac{∂f_2}{∂x_2} = cos(x_2) \space\space \dfrac{∂f_2}{∂x_3} = -1$ 
## $\dfrac{∂f_1}{∂x_1} = x_2e^{x_1x_2} \space\space \dfrac{∂f_2}{∂x_2} = x_1e^{x_1x_2} \space\space \dfrac{∂f_3}{∂x_3} = 2x_3$    
## $
J(x_i) = 
\left(\begin{array}{cc} 
log(x_2) & \dfrac {x_1}{x_2} + x_3 & x_2\\
2x_1 & cos(x_2) & -1\\
x_2e^{x_1x_2} & x_1e^{x_1x_2} & 2x_3
\end{array}\right)
$

### Definimos cada $x_i$ y el vector $X_0$
## $x_1 = 1; x_2 = 1; x_3 = 1$
## $X_0 = (1, 1, 1)$

## Se emplea un vector $ε = (ε_1, ε_2, ε_3)$ para almacenar errores relativos, con cada elemento definido como: $ε_i = \biggl|\dfrac {Δx_i}{x_i}\biggr|$

```python
# Encontraremos un detalle respecto a estos valores
x1 = 1
x2 = 1
x3 = 1
X = [x1, x2, x3]

E = [100, 100, 100]
for a in range(100):
    print("Iteración:", a)
    print(sum(np.array(E)))
    if all(e < 0.001 for e in E):
        print(a+1, "iteraciones para todo error RP < 0.001%")
        break
    
    # Buscamos obtener: delta_x=-J(x_i)^-1*F(x_i)
    
    # Conseguimos -F(x_i)
    F1 = x1*log(x2) + x2*x3 - 3
    F2 = (x1**2) + sin(x2) - x3 - 1
    F3 = (e**(x1*x2)) + (x3**2) - 4
    
    # Detalle: aquí cada una entra negativa
    # porque el método iguala a -F(x_i)
    F = np.array([F1, F2, F3])
    
    # Obtenemos J(x_i)
    dF1dx1 = log(x2)
    dF1dx2 = (x1/x2) + x3
    dF1dx3 = x2

    dF2dx1 = 2*x1
    dF2dx2 = cos(x2)
    dF2dx3 = -1

    dF3dx1 = x2*(e**(x1*x2))
    dF3dx2 = x1*(e**(x1*x2))
    dF3dx3 = 2*x3

    J = [ [dF1dx1, dF1dx2, dF1dx3],
          [dF2dx1, dF2dx2, dF2dx3],
          [dF3dx1, dF3dx2, dF3dx3] ]
    
    J_inv = np.linalg.inv(J)
    
    # obtener delta_x
    d1, d2, d3 = - J_inv @ F
    D = [d1, d2, d3]
    
    # Cálculo de Errores
    for i in range(len(D)):
        E[i] = abs(D[i]/X[i]) * 100
    
    # Ajustamos este cambio en las x ya que calculamos una diferencia
    x1 += d1
    x2 += d2
    x3 += d3
    
    X = [x1, x2, x3]
    print(f"x1: {round(x1, 10)}, x2: {round(x2, 10)}, x3: {round(x3, 10)}")
    print("")
```

# ¿Qué ocurre?
## Al ser el método muy susceptible a sus condiciones iniciales; es muy sencillo encontrar problemas si las funciones dentro de F(X) tienen dominios limitados (e.g. para $ln(x); \space D=\{x|x>0\}$)

## Para solventar esto, se probarán valores enteros para cada variable dentro del intervalo [-5, 5] 

## Además, se empleara la ecuación vectorial $\Delta x=-J(x_i)^{-1}F(x_i)$

```python
def input_functions(n):
    functions = []
    for i in range(n):
        func_str = input(f"Introduce la función f_{i+1}(x1, ..., xn): ")
        functions.append(eval(f"lambda x: {func_str}"))
    return functions

def evaluate_functions(functions, X):
    x1, x2, x3 = X
    return np.array([func(x1, x2, x3) for func in functions], dtype=float)

def jacobian_matrix(functions, x, h=1e-8):
    n = len(x)
    J = np.zeros((n, n))
    for i in range(n):
        x_h = np.copy(x)
        x_h[i] += h
        f_x_h = evaluate_functions(functions, x_h)
        f_x = evaluate_functions(functions, x)
        J[:, i] = (f_x_h - f_x) / h
    return J

def newton_method_general(functions, x0, tol=1e-5, max_iter=100):
    X = np.array(x0, dtype=float)
#     print(f"{'Iteración':<10}{'x':<20}{'Δx':<20}{'Error Relativo (%)':<20}")
    for k in range(max_iter):
        Fx = evaluate_functions(functions, X)
        Jx = jacobian_matrix(functions, X)
        Jx_inv = np.linalg.inv(Jx)
        delta_x = -Jx_inv @ Fx
        x_new = X + delta_x
        error_relativo = np.linalg.norm(delta_x / (X + 1e-10)) * 100  # Evitar división por cero
#         print("")
#         print(f"{k:<10}{X[0]:<20.10f}{delta_x[0]:<20.10f}{error_relativo:<20.10f}")
#         print(f"{k:<10}{X[1]:<20.10f}{delta_x[1]:<20.10f}")
#         print(f"{k:<10}{X[2]:<20.10f}{delta_x[2]:<20.10f}")
        X = x_new
        if error_relativo < tol:
            break
            
    try:    
        return tuple([round(X[0], 10), round(X[1], 10), round(X[2], 10)])
    except:
        return None
```

```python
F1 = lambda x1, x2, x3: x1*log(x2) + x2*x3 - 3
F2 = lambda x1, x2, x3: (x1**2) + sin(x2) - x3 - 1
F3 = lambda x1, x2, x3: (e**(x1*x2)) + (x3**2) - 4
functions = [F1, F2, F3]

soluciones_puntos_i = defaultdict()

check_range = np.arange(-8, 9)
for x1 in check_range:
    for x2 in check_range:
        for x3 in check_range:
            x0 = np.array([x1, x2, x3])
            try: 
                solution = newton_method_general(functions, x0)
                if (any(np.isnan(solution))):
                    continue
                if solution in soluciones_puntos_i.keys():
                    soluciones_puntos_i[solution].append(x0)
                if solution not in soluciones_puntos_i.keys():
                    soluciones_puntos_i[solution] = [x0]
            except: 
                pass
```
